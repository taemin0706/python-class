# Class
1. User 클래스 생성
 - 생성자 함수는 유저 정보 입력(이름, 휴대폰번호, 생년월일)
 - info() 함수를 생성 : 유저의 정보를 출력
2. Wallet 클래스 생성
 - User를 상속
 - 생성자 함수는 소유자의 정보(이름, 휴대폰번호, 생년월일), 지갑 브랜드, 지갑의 가치, 잔액(0) 입력
 - work() 함수를 생성 : 일의 종류는 "TM", "Alba", "Serving"이고 TM을 하는 경우 지갑의 잔액이 5000 증가, "Alba"를 하는 경우 잔액 10000 증가, Serving을 하는 경우 잔액이 15000 증가
 - wallet_info() 함수를 생성 : 지갑의 소유자 정보와 잔액을 출력

```python
# User 클래스 생성
class User : 
    # 생성자 함수인 __init__ 생성
    def __init__(self, _name, _phone, _birth) :
        self.name = _name
        self.phone = _phone
        self.birth = _birth
    # 유저의 정보를 출력하는 함수
    def info(self) :
        self.result = [self.name, self.phone, self.birth]
        return self.result      
```

```python
user1 = User("test", "01012345678", "000101")
user2 = User("test2", "01098765432", "850101")
```

```python
print(user1.info())  # 인자값을 안넣는 이유는 info(self)의 self에 user1, user2가 들어가기 때문이다.
print(user2.info())
```

```python
# Wallet 클래스 선언
class Wallet(User) :
    # 생성자 함수 생성
    def __init__(self, _name, _phone, _birth, _brand, _value, _balance = 0) :
        # 부모 클래스의 생성자 함수를 실행
        super().__init__(_name, _phone, _birth)
        self.brand = _brand
        self.value = _value
        self.balance = _balance
    
    # work() 함수 생성 -> 일의 종류
    def work(self, _work_type) : 
        # 업무별 금액 설정
        if _work_type == "TM" :
            self.cost = 5000
        elif _work_type == "Alba" :
            self.cost = 10000
        elif _work_type == "Serving" :
            self.cost = 15000
        else :
            return "work_type이 맞지 않습니다"
        
        # 잔액 변경
        self.balance += self.cost
        return f"{_work_type} 일을 성공하였습니다. 현재 잔액은 {self.balance}입니다."
    
    # 지갑의 정보를 출력하는 함수
    def wallet_info(self) :
        print(f"지갑의 소유자는 {self.name}")
        print(f"현재 잔액은 {self.balance}")
    
    # 오버라이딩 - 부모 클래스의 기능을 변경하여 사용
    def info(self) : 
        print(f"소유자의 이름은 {self.name}")
        print(f"소유자의 브랜드는 {self.brand}")
        print(f"소유자의 가치는 {self.value}")
```

```python
wallet1 = Wallet("test", "01012345678", "000101", "gucci", 1000000)
wallet2 = Wallet("test2", "01098765432", "850101", "chanel", 5000000)
```
